* The [beer foam dataset](http://www.stat.ufl.edu/~winner/datasets.html) was collected by A. Leike and published in their work titled "Demonstration of the Exponential Decay Law Using Beer Froth" in 2002.

#### Data Set Description

The data contains measurements of wet foam height and beer height at various time points for 3 brands of beer. The author of this data set fit an *exponential decay model* of the form $H(t) = H_0 e^{-\lambda t}$.

The data set is saved as a *.csv* file (**"beer_foam.csv"**) with information about the foam height (in cm) from 3 brands of beer over 15 measurement times (in seconds) after the time of pour.

The file is organized in 4 columns:
1. Time from pour (in seconds)
2. Erdinger Weissbier foam height (in cm)
3. Augustinerbrau Munchen foam height (in cm)
4. Budweiser foam height (in cm)

## Objectives

* Implement polynomial linear regression model on a given dataset
* Implement exponential regression model on a given dataset
* Utilize the train model to make predictions
* Assess the goodness-of-fit of each model
* Make observations and discuss limitations of model design in forecasting

1. Load the data using ```pandas```.

Consider the first 12 samples the training set, and the last 3 samples the test set.

2. Build and train a polynomial regression model for **each** bear brand with model order $M=4$.

3. Build and train an exponential model with 2 parameters for **each** bear brand.

4. Predict the foam height for **each** beer brand for $t=450~$seconds after pour using the trained polynomial regression model (from problem 2) and exponential model (from problem 3).

5. Compare both models using plots (qualitative measure) and select a measure to assess the goodness-of-fit (quantitative measure). Based on these results and prediction for $t=450~$seconds for both models, discuss which performed best on this data.

**Problem 1: Loading data using pandas**

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
````

````python
data = pd.read_csv('beer_foam.csv')
data
````

````python
D = data.to_numpy()
D
````

````python
xtr = D[0:12,0] # training data - time
tr1 = D[0:12,1] # training data - Erdinger
tr2 = D[0:12,2] # training data - Augustinerbrau
tr3 = D[0:12,3] # training data - Budweiser

xte = D[12:15,0] # test data - time
te1 = D[12:15,1] # test data - Erdinger
te2 = D[12:15,2] # test data - Augustinerbrau
te3 = D[12:15,3] # test data - Budweiser

plt.figure(figsize=(10,7))
plt.scatter(xtr,tr1, label = 'Training Data Erdinger')
plt.scatter(xtr,tr2, label = 'Training Data Augustinerbrau')
plt.scatter(xtr,tr3, label = 'Training Data Budweiser')
plt.scatter(xte,te1, marker = '*', label = 'Test Data Erdinger')
plt.scatter(xte,te2, marker = '*', label = 'Test Data Augustinerbrau')
plt.scatter(xte,te3, marker = '*', label = 'Test Data Budweiser')
plt.legend(loc='upper right')
plt.legend(fontsize=15);
````

**Problem 2: Polynomial Regression Model**

````python
def PolynomialRegression(x,t,M):
    X = np.array([x**m for m in range(M+1)]).T
    w = np.linalg.inv(X.T@X)@X.T@t
    y = X@w
    error = t-y
    return w, y, error

def PolynomialRegression_test(x,M,w):
    X = np.array([x**m for m in range(M+1)]).T 
    y = X@w
    return y
````

````python
Mpoly = 4 # polynomial model order
````

````python
#Erdinger
w1, ytr1, err1 = PolynomialRegression(xtr,tr1,Mpoly)
y1_test = PolynomialRegression_test(xte, Mpoly, w1)

plt.figure(figsize=(10,7))
plt.scatter(xtr,tr1, label = 'Training Data')
plt.plot(xtr,ytr1, 'r', label = 'Estimated Polynomial')
plt.scatter(xte,te1, marker = '*', label = 'Test Data')
plt.plot(xte,y1_test, 'k--', label = 'Predictions' )
plt.legend(loc='upper right')
plt.title('Erdinger Foam Height', size = 15)
plt.xlabel('time(s)', size = 15)
plt.ylabel('Foam height (cm)', size = 15)
plt.legend(fontsize=15);
````

````python
#Augustinerbrau
w2, ytr2, err2 = PolynomialRegression(xtr,tr2,Mpoly)
y2_test = PolynomialRegression_test(xte, Mpoly, w2)

plt.figure(figsize=(10,7))
plt.scatter(xtr,tr2, label = 'Training Data')
plt.plot(xtr,ytr2, 'r', label = 'Estimated Polynomial')
plt.scatter(xte,te2, marker = '*', label = 'Test Data')
plt.plot(xte,y2_test, 'k--', label = 'Predictions' )
plt.legend(loc='upper right')
plt.title('Augustinerbrau Foam Height', size = 15)
plt.xlabel('time(s)', size = 15)
plt.ylabel('Foam height (cm)', size = 15)
plt.legend(fontsize=15);
````

````python
#Budweiser
w3, ytr3, err3 = PolynomialRegression(xtr,tr3,Mpoly)
y3_test = PolynomialRegression_test(xte, Mpoly, w3)

plt.figure(figsize=(10,7))
plt.scatter(xtr,tr3, label = 'Training Data')
plt.plot(xtr,ytr3, 'r', label = 'Estimated Polynomial')
plt.scatter(xte,te3, marker = '*', label = 'Test Data')
plt.plot(xte,y3_test, 'k--', label = 'Predictions' )
plt.legend(loc='upper right')
plt.title('Budweiser Foam Height', size = 15)
plt.xlabel('time(s)', size = 15)
plt.ylabel('Foam height (cm)', size = 15)
plt.legend(fontsize=15);
````

**Problem 3: Exponential Model**

````python
#Erdinger

Mexp=1

tr1_log = np.log(tr1)

w1_log, y1_log, e1_log = PolynomialRegression(xtr, tr1_log, Mexp)

pred_train_exp1 = np.exp(w1_log[0] + w1_log[1]*xtr)
pred_test_exp1 = np.exp(w1_log[0]+w1_log[1]*xte)

plt.figure(figsize=(10,7))
plt.scatter(xtr, tr1, label='Training Data')
plt.plot(xtr, pred_train_exp1,'r', label='Estimated Exponential Model')
plt.scatter(xte, te1, marker = '*', label = 'Test Data')
plt.plot(xte, pred_test_exp1, 'k--', label = 'Predictions' )
plt.legend(fontsize=15)
plt.title('Erdinger Foam Height', size = 15)
plt.xlabel('time(s)', size = 15)
plt.ylabel('Foam height (cm)', size = 15)
plt.legend(fontsize=15);
````

````python
#Augustinerbrau

tr2_log = np.log(tr2)

w2_log, y2_log, e2_log = PolynomialRegression(xtr, tr2_log, Mexp)

pred_train_exp2 = np.exp(w2_log[0] + w2_log[1]*xtr)
pred_test_exp2 = np.exp(w2_log[0]+w2_log[1]*xte)

plt.figure(figsize=(10,7))
plt.scatter(xtr, tr2, label='Training')
plt.plot(xtr, pred_train_exp2,'r', label='Trained')
plt.scatter(xte, te2, marker = '*', label = 'Test Data')
plt.plot(xte, pred_test_exp2, 'k--', label = 'Predictions' )
plt.legend(fontsize=15)
plt.title('Augustinerbrau Foam Height', size = 15)
plt.xlabel('time(s)', size = 15)
plt.ylabel('Foam height (cm)', size = 15)
plt.legend(fontsize=15);
````

````python
#Budweiser

tr3_log = np.log(tr3)

w3_log, y3_log, e3_log = PolynomialRegression(xtr, tr3_log, Mexp)

pred_train_exp3 = np.exp(w3_log[0] + w3_log[1]*xtr)
pred_test_exp3 = np.exp(w3_log[0]+w3_log[1]*xte)

plt.figure(figsize=(10,7))
plt.scatter(xtr, tr3, label='Training')
plt.plot(xtr, pred_train_exp3,'r', label='Trained')
plt.scatter(xte, te3, marker = '*', label = 'Test Data')
plt.plot(xte, pred_test_exp3, 'k--', label = 'Predictions' )
plt.legend(fontsize=15)
plt.title('Budweiser Foam Height', size = 15)
plt.xlabel('time(s)', size = 15)
plt.ylabel('Foam height (cm)', size = 15)
plt.legend(fontsize=15);
````

````python
#Erdinger coefficient of determination

import scipy.stats as stats

print('POLYNOMIAL MODEL')
_, _, r, _, _ = stats.linregress(np.sort(te1), np.sort(y1_test))
print('Coefficient of Determination: ',r**2)
print('-------------------')
print('EXPONENTIAL MODEL')
_, _, r, _, _ = stats.linregress(np.sort(te1), np.sort(pred_test_exp1))
print('Coefficient of Determination: ',r**2)
````

We can also observe that the test results of exponential model follow the training data which is not the case with polynomial model. The polynomial model fits the data well for M = 3 but overfits for M = 4 and above.

2. This can be confirmed by the calculated coefficients of determination for all three beer brand for both the models.

3. Choosing the right type of model has major implications on how it performs with data outside of the training set. Here, both polynomial and exponential fit the training data but has difference in performance when it comes to test data. Hence, choosing a model is a crucial part in machine learning algorithms. Qualitative and Quantitave assessments assist us with choosing the right type of model for a given dataset.
